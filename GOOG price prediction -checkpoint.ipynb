# In this we created an Artificial Neural Network called Long Short Term Memory to predict the future price of Stock.

# Imported Libraries :-

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_datareader as web
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,LSTM
plt.style.use('fivethirtyeight')
```

# Getting GOOGLE Stock Details from Online resource YAHOO from "2012-01-01" to "2019-12-18" to predict it's future price.

```python
df=web.DataReader('goog',data_source='yahoo',start='2012-01-01',end='2019-12-18')
df.head() # It will show top 5 data
```

# Finding Correlation of Dataset using df.corr()

```python
df.corr()
```

# Visualizing correlation in heatmap

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

# To know the shape of dataset (rows and columns) we are using df.shape

```python
df.shape # 2004 rows and 6 columns
```

# Vizualizing the Closing Price present in Dataset 

```python
plt.figure(figsize=(16,8))
plt.title('close price history')
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=19)
plt.ylabel('close Price USD ($)',fontsize=19)
plt.show()
```

#  Created DataFrame that contains only closing price to train the model and convert the dataframe into numpy array.


# Our Train_data have length of 1604


```python
data=df.filter(['Close']) # contain only Close Price column
#converting the dataframe to a numpy
dataset=data.values
#getting no of raws to train the model
train_data_len=math.ceil(len(dataset)*.8)  #.8 for rounding up data value example 2.883=>3

train_data_len
```

# Scaling the data into MinMaxScalar to make easy while training the dataset.

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(dataset)
```

# Spliting the dataset into train_x and train_y to read past 60th data.


# where train_x contains the value from 0-59 and train_y contains values from 61.

```python
#creating training dataset
#creating scaled training dataset
train_data=scaled_data[0:train_data_len,:]
#spliting the data into x_train and y_train dataset
train_x=[]
train_y=[]

for i in range(60,len(train_data)):
    train_x.append(train_data[i-60:i,0])#append values from 1 to 60 in 0th column(0-59) 
    train_y.append(train_data[i,0])#contains 61th value
    if i <=61:
        print(train_x)
        print(train_y)
```

# converting train_x & train_y into numpy array for lstm model, 
# and reshape it into 3d array.

```python
train_x,train_y=np.array(train_x),np.array(train_y)
```

```python
#reshaping the data(bcz lstm need 3d data)
train_x=np.reshape(train_x,(train_x.shape[0],train_x.shape[1],1))#train_x.shape[0] contains(584) and train_x.shape[1] contains(60)
train_x.shape
```

# We are using LSTM model for creating 50 neurons to train our model.

# Here Sequential() model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one output tensor.

```python
# creating lstm model
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(train_x.shape[1],1)))
model.add(LSTM(50,return_sequences=False))# False because we are not taking anymore lstm layer to train model
model.add(Dense(25))
model.add(Dense(1))
```

# Compiling the model to define the Loss Function.

# For that we are using "Adam optimizer" which is a replacement optimization algorithm for stochastic gradient descent for training deep learning models.

```python
# compiling the model
model.compile(optimizer='adam',loss="mean_squared_error") # loss->MSE for how well the model is trained.
```

# Training the model in batches wrt. no of iterations by using EPOCH

```python
model.fit(train_x,train_y,batch_size=1,epochs=1)# Epochs -> no. of iteration to train the model.
```

# Creating the testing dataset
# Where the new array having scaled values from index 1544 to 2004

```python
data_test=scaled_data[train_data_len-60:,:] # spliting datset from train_data_len to less than 60 to rest of the data.
#creating the data set test_x and test_y
test_x=[]
test_y=dataset[train_data_len:,:]
for i in range(60,len(data_test)):
    test_x.append(data_test[i-60:i,0])
```

```python
#converting data to a numpy array
test_x=np.array(test_x)
```

```python
#reshaping the data into 3d 
test_x=np.reshape(test_x,(test_x.shape[0],test_x.shape[1],1))
```

# Getting the models that predict the price value.

```python
# model that have predicted value price 
prediction=model.predict(test_x)
prediction=scaler.inverse_transform(prediction) # unscaling the dataset to test_y
```

# ROOT_MEAN_SQUARE_ERROR that tells the performance of the model.

```python
#getting the RMSE to know how the model perform
rmse=np.sqrt(np.mean(prediction - test_y)**2)
rmse
```

# Visualizing the graphs to compare the 'actual closing price' and 'predicted price' of the stock.

```python
#plotting the data
train=data[:train_data_len]
valid=data[train_data_len:]
valid['prediction']=prediction
#visualizing
plt.figure(figsize=(16,8))
plt.title('Model')
plt.xlabel('Date',fontsize=19)
plt.ylabel('Close Price USD ($)',fontsize=19)
plt.plot(train['Close'])
plt.plot(valid[['Close','prediction']]) # for actual price and predicted price
plt.legend(['Train','Validation','Prediction'],loc='lower right')
plt.show()
```
