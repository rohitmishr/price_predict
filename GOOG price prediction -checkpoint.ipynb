```python
#import Libraries
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pandas_datareader as web
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense,LSTM
plt.style.use('fivethirtyeight')
```

```python
#getting the stock details
df=web.DataReader('goog',data_source='yahoo',start='2012-01-01',end='2020-10-10')
df.head()
```

```python
#getting raws and columns of dataset
df.shape
```

```python
#vizualizing the closing price
plt.figure(figsize=(16,8))
plt.title('close price history')
plt.plot(df['Close'])
plt.xlabel('Date',fontsize=19)
plt.ylabel('close Price USD ($)',fontsize=19)
plt.show()
```

```python
#creating new dataframe only with close column
data=df.filter(['Close'])
#converting the dataframe to a numpy
dataset=data.values
#getting no of raws to train the model
train_data_len=math.ceil(len(dataset)*.8)

train_data_len
```

```python
#scaling the data
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(dataset)

scaled_data
```

```python
#creating training dataset
#creating scaled training dataset
train_data=scaled_data[0:train_data_len,:]
#spliting the data into x_train and y_train dataset
train_x=[]
train_y=[]

for i in range(60,len(train_data)):
    train_x.append(train_data[i-60:i,0])
    train_y.append(train_data[i,0])
    if i <=61:
        print(train_x)
        print(train_y)
```

```python
#converting train_x & train_y into numpy array
train_x,train_y=np.array(train_x),np.array(train_y)
```

```python
#reshaping the data(bcz lstm need 3d data)
train_x=np.reshape(train_x,(train_x.shape[0],train_x.shape[1],1))
train_x.shape
```

```python
# creating lstm model
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(train_x.shape[1],1)))
model.add(LSTM(50,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
# compiling the model
model.compile(optimizer='adam',loss="mean_squared_error")
```

```python
#training the model
model.fit(train_x,train_y,batch_size=1,epochs=1)
```

```python
#creating the testing dataset
#creating the new darra having scaled values from index 1543 to 2003
data_test=scaled_data[train_data_len-60:,:]
#creating the data set test_x and test_y
test_x=[]
test_y=dataset[train_data_len:,:]
for i in range(60,len(data_test)):
    test_x.append(data_test[i-60:i,0])
```

```python
#converting data to a numpy array
test_x=np.array(test_x)
```

```python
#reshaping the data into 3d 
test_x=np.reshape(test_x,(test_x.shape[0],test_x.shape[1],1))
```

```python
#getting models predicted value price
prediction=model.predict(test_x)
prediction=scaler.inverse_transform(prediction)
```

```python
#getting the RMSE
rmse=np.sqrt(np.mean(prediction - test_y)**2)
rmse
```

```python
#plotting the data
train=data[:train_data_len]
valid=data[train_data_len:]
valid['prediction']=prediction
#visualizing
plt.figure(figsize=(16,8))
plt.title('MOdel')
plt.xlabel('Date',fontsize=19)
plt.ylabel('Close Price USD ($)',fontsize=19)
plt.plot(train['Close'])
plt.plot(valid[['Close','prediction']])
plt.legend(['Train','Val','Prediction'],loc='lower right')
plt.show()
```
